```python
df=pd.read_csv('mbti_1.csv')
df
```

```python
df.info()
```

```python
x = df['posts']
y = df['type']

x.shape, y.shape
```

```python
count_vec= CountVectorizer(ngram_range=(1,4),max_features=10000)
```

```python
t=count_vec.fit(x.values)
voc=count_vec.vocabulary_
sort_voc1=dict(sorted(voc.items(), key=lambda item: item[1]))
print(sort_voc1)
```

```python
def fun(a):
    if a>=1:
        print(a)
        return fun(a-1)
    else: 
        print(a)
        return 1
fun(3)
```

```python
mm=t.fit_transform(x).toarray()
mm
```

```python
x_train,x_test, y_train, y_test = train_test_split(mm, y, random_state=0)
```

```python
%%time

knn = KNeighborsClassifier(n_neighbors=3, metric='manhattan')
knn.fit(x_train,y_train)
knn.score(x_train, y_train),knn.score(x_test, y_test)
```

```python
%%time

svm = SVC(kernel='sigmoid', C=1000)
svm.fit(x_train,y_train)
svm.score(x_train, y_train),svm.score(x_test, y_test)
```

```python
%%time

nb =GaussianNB()
nb.fit(x_train,y_train)
nb.score(x_train, y_train),nb.score(x_test, y_test)
```

```python
dt = DecisionTreeClassifier(criterion='entropy', max_depth=15)
dt.fit(x_train,y_train)
dt.score(x_train, y_train),dt.score(x_test, y_test)
```
